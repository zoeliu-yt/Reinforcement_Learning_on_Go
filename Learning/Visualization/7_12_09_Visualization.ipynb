```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Evaluation
## Load Data 
Random, Q_learning and Policy 

```python
df = pd.read_csv('PG.csv')
pg = df['0']
df = pd.read_csv('Q.csv')
q = df['0']
df = pd.read_csv('Random.csv')
r = df['0']
len(r)
```

## Skip Sampling

In every 4 consecutive samples, draw one. Therefore all samples are approximately independent.
more info

```python
skip = 4
pg_samples = []
q_samples = []
r_samples = []
for i in range(len(pg)):
        if i % skip == 0:
            pg_samples.append(pg[i])
            q_samples.append(q[i])
            r_samples.append(r[i])
len(r_samples)
```

## Density Plot


```python
j = 0
bk = 200
k = 1 
for i in range(len(r_samples)):
    if (i+1) % bk == 0:
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex='all')
        ax1.hist(r_samples[j:i], color = 'blue', edgecolor = 'black',
                 bins = int(180/5))
        
        
        ax2.hist(q_samples[j:i], color = 'blue', edgecolor = 'black',
                 bins = int(180/5))
        
        
        ax3.hist(pg_samples[j:i], color = 'blue', edgecolor = 'black',
                 bins = int(180/5))
        #plt.ylim(ymin=0, ymax = 30)
        #plt.xlim(xmin=0, xmax = 200)
        #plt.subplots(1, 1, sharex='col')
        #plt.subplots(1, 1, sharex='row')
        
        #plt.xlabel('reward')
        #plt.ylabel('no.episode')
        plt.title('Iteration:'+str(k)+'-'+str((i+1)*4))
        plt.setp([ax1], xlabel='random reward')
        plt.setp([ax2], xlabel='Q-learning reward')
        plt.setp([ax3], xlabel='Policy Gradient reward')
        plt.setp([ax1], ylabel='no.episode')
        #plt.setp(fig, title='Iteration:'+str(k)+'-'+str((i+1)*4))
        plt.show()
        k = (i+1)*4+1
        j = i
```

```python
j = 0
bk = 200
k = 1 
for i in range(len(r_samples)):
    if (i+1) % bk == 0:
        print('Iteration:'+str(k)+'-'+str((i+1)*4))
        print('Random:', np.mean(r_samples[j:i]), 'Q-learning:', np.mean(q_samples[j:i]), 'Policy Gradient:',np.mean(pg_samples[j:i]))
        print('Random:', np.var(r_samples[j:i]), 'Q-learning:', np.var(q_samples[j:i]), 'Policy Gradient:',np.var(pg_samples[j:i]))
```
