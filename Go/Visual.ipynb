# Eposodes

```python

a = np.load("non.npy")
b = np.load("100.npy")
c = np.load("1000.npy")
d = np.load("10000.npy")
```

```python
tie = [list(a).count(0)]
win = [list(a).count(1)]
lose = [list(a).count(-1)]
for i in b:
    tie.append(list(i).count(0))
    win.append(list(i).count(1))
    lose.append(list(i).count(-1))
for i in c:
    tie.append(list(i).count(0))
    win.append(list(i).count(1))
    lose.append(list(i).count(-1))
for i in d:
    tie.append(list(i).count(0))
    win.append(list(i).count(1))
    lose.append(list(i).count(-1))
```

```python
tie = tie[0:len(tie)-7]
win = win[0:len(win)-7]
lose = lose[0:len(lose)-7]
```

```python
x = list(np.linspace(0, 1000, 11)) + list(np.linspace(2000, 10000, 9))+list(np.linspace(20000, 50000, 4))
```

```python

y = win 

plt.plot(x, np.asarray(y)/1000)
plt.xlabel('Episodes')
plt.ylabel('Win Rate')
plt.show()
```

```python

y = tie 

plt.plot(x, np.asarray(y)/1000)
plt.xlabel('Episodes')
plt.ylabel('Tie Rate')
plt.show()
```

```python

y = lose 

plt.plot(x, np.asarray(y)/1000)
plt.xlabel('Episodes')
plt.ylabel('Lose Rate')
plt.show()
```

```python
ls = [list(a)]

for i in b:
    ls.append(i)
for i in c:
    ls.append(i)
for i in d:
    ls.append(i)
```

```python
from scipy.stats import iqr
import scipy
def es(array):
    array = np.asarray(array)
    value_at_risk = scipy.stats.scoreatpercentile(array, 100 * (1-0.05))
    cvar = array[array <= value_at_risk].mean()
    return cvar
```

```python
es_ls = []
iqr_ls = []
for i in ls:
    es_ls.append(es(i))
    iqr_ls.append(iqr(i))
es_ls = es_ls[0:len(es_ls)-7]
iqr_ls = iqr_ls[0:len(iqr_ls)-7]
es_ls
```

```python
import pandas as pd
df = pd.DataFrame({"Num_of_Episodes": x, 
                   "Win_Rate": np.asarray(win)/1000, 
                   "Lose_Rate": np.asarray(lose)/1000, 
                   "Tie_Rate": np.asarray(tie)/1000,
                   "IQR": iqr_ls,
                   "CVaRs": es_ls})
df
```
